```python
import seaborn as sns


# 1. Isolate legitimate and fraud transactions
legit_df = df[df['Class'] == 0].drop(['Class'], axis=1)
fraud_df = df[df['Class'] == 1].drop(['Class'], axis=1)

# 2. Set up the matplotlib figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 10))

# 3. Plot Legitimate Heatmap (Will look almost completely blank/zero)
sns.heatmap(legit_df.corr(), cmap='coolwarm', vmin=-1, vmax=1, ax=ax1, cbar=False)
ax1.set_title('Legitimate Transactions Correlation\n(PCA forces this to be near 0)', fontsize=14)

# 4. Plot Fraud Heatmap (Will show distinct blocks of correlation)
sns.heatmap(fraud_df.corr(), cmap='coolwarm', vmin=-1, vmax=1, ax=ax2)
ax2.set_title('Fraudulent Transactions Correlation\n(Hidden patterns reveal themselves!)', fontsize=14)

plt.tight_layout()
```

#### As the feautres have been selected through pca so the first graph shows that they are independent and not related when fraud doesnt occur
#### Second graph shows the relation when there is fraud occuring

```python
# Now checking for class imbalance
# Calculate counts and percentages
imbalance_counts = df['Class'].value_counts()
imbalance_pct = df['Class'].value_counts(normalize=True) * 100

print("--- Class Distribution ---")
print(f"Legitimate Transactions (Class 0): {imbalance_counts[0]} ({imbalance_pct[0]:.3f}%)")
print(f"Fraudulent Transactions (Class 1): {imbalance_counts[1]} ({imbalance_pct[1]:.3f}%)")
```

#### Here we can see that there is a humungous class imbalance so we have to deal with it
